# Vehicle Collision Data Visualization 

```python
import matplotlib.pyplot as plt
import pandas as pd

crashes = pd.read_csv('data/clean_data/clean_data_main.csv')
cars = pd.read_csv('data/clean_data/clean_data_car.csv')
non_covid = crashes.query("'2019-03-16' <= CollisionDate <= '2019-12-31'")
covid = crashes.query("'2020-03-16' <= CollisionDate <= '2020-12-31'")
```

Deadliest Roads

```python
crashes.groupby('RoadwayName')['NumberKilled'].sum().sort_values(ascending=False).head(10).plot(kind='barh', y='RoadwayName', xlabel='Deaths', ylabel='Road')
```

Number of collisions before covid vs during covid

```python
total_nc = non_covid['Index'].count()
total_c = covid['Index'].count()
change = pd.DataFrame({'Collisions':[total_nc, total_c]},
                       index=['Before Covid', 'During Covid'])
change.plot(kind='pie', y=0)
```

Number of collisions before covid dropped most during weekend

```python
non_covid['DayofWeek'].value_counts().plot(kind='bar', xlabel='Day', ylabel='Collisions')

```

Number of collisions during covid more consistent from day to day with less of a drop off during weekend

```python
covid['DayofWeek'].value_counts().plot(kind='bar', xlabel='Day', ylabel='Collisions')
```

Number of collisions by vehicle model

```python
cars.groupby([ 'ModelDescription']).size().sort_values(ascending=False).head(5).plot(kind='line', xlabel='Car Model', ylabel='Collisions')
```
